## Outlier Treatment

```python
plt.figure(figsize=(10,10))
df.boxplot(vert=0)
plt.title('Outliers', fontsize=16)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3-Q1
lower_range = Q1-(1.5*IQR)
upper_range = Q3+(1.5*IQR)
```

```python
print("Upper Range:")
print(upper_range)
print()
print("Max_ Age", df['age'].max())
print("Max Duration of pitch", df['durationofpitch'].max())
print("Max Number of trips", df['numberoftrips'].max())
print("max monthly income", df['monthlyincome'].max())
```

```python
print("Lower Range:")
print(lower_range)
print()
print("min Age", df['age'].min())
print("min Duration of pitch", df['durationofpitch'].min())
print("min Number of trips", df['numberoftrips'].min())
print("min monthly income", df['monthlyincome'].min())
```

```python
df['monthlyincome'] = np.where(df['monthlyincome'] > 32760.75, 32760.75, df['monthlyincome'])
df['monthlyincome'] = np.where(df['monthlyincome'] < 13102.75, 13102.75, df['monthlyincome'])

df['durationofpitch'] = np.where(df['durationofpitch'] > 34, 34, df['durationofpitch'])

df['numberoftrips'] = np.where(df['numberoftrips'] > 7, 7, df['numberoftrips'])

```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score, f1_score, accuracy_score, roc_auc_score
```

```python
x = df.drop(['prodtaken','clusters'], axis=1)
y = df['prodtaken']
```

```python
from sklearn.model_selection import train_test_split
tts = train_test_split
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size = 0.30, random_state=42)
```

```python
rfcl = RandomForestClassifier(n_estimators=150, random_state=42)
rfcl = rfcl.fit(x_train, y_train)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

### Predicting Train and Test data

```python
rfcl
```

```python
ytrain_predict = rfcl.predict(x_train)
ytest_predict = rfcl.predict(x_test)
```

```python
ytrain_predict.shape
```

```python
ytest_predict.shape
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
confusion_matrix(y_train, ytrain_predict)
#sns.heatmap(confusion_matrix(y_train, ytrain_predict), annot=True, fmt='d', cbar=False, cmap='rainbow')
```

```python
cm = confusion_matrix(ytrain_predict,y_train, labels=rfcl.classes_)


disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=rfcl.classes_)
disp.plot()
```

```python
rftrain_recall = recall_score(y_train, ytrain_predict)
print('Training Recall', rftrain_recall)
rftrain_roc_auc = roc_auc_score(y_train,rfcl.predict_proba(x_train)[:,1])
print('Training roc_auc',rftrain_roc_auc)
```

```python
print(classification_report(y_train, ytrain_predict))
```

```python
confusion_matrix(y_test, ytest_predict)
cm = confusion_matrix(y_test, ytest_predict, labels=rfcl.classes_)


disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=rfcl.classes_)
disp.plot()
```

```python
print(classification_report(y_test, ytest_predict))
```

```python
rftest_recall = recall_score(y_test, ytest_predict)
print('Test Recall', rftest_recall)
rftest_roc_auc = roc_auc_score(y_test,rfcl.predict_proba(x_test)[:,1])
print('Test roc_auc',rftest_roc_auc)
```

```python
metrics_df = pd.DataFrame({'Train_Recall': rftrain_recall, 'Test_Recall': rftest_recall, 'Train_roc_auc': rftrain_roc_auc, 'Test roc_auc': rftest_roc_auc}, index=['Random Forest'])
metrics_df
```
